```python
#W5A8
import math
def Check(s1, s2):
    res = abs(len(s1) - len(s2))
    for i in range (0, min(len(s1), len(s2))):
        if(s1[i] != s2[i]):
            res += 1
    return res

a, b = map(int, input().split())
s1, s2 = '', ''

while(a > 0):
    s1 = str(a % 2) + s1
    a //= 2

while(b > 0):
    s2 = str(b % 2) + s2
    b //= 2
print(Check(s1, s2))
```

```python
#W5A10
def Check(a, b):
    char = {}
    if(len(a) != len(b)):
        return False
    for i in range (len(a)):
        if(a[i] in char):
            if(char[a[i]] != b[i]):
                return False
        else:
            char[a[i]] = b[i]
    return True
    
a = str(input())
b = str(input())
print(Check(a, b))
```
